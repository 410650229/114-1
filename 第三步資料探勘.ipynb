```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 匯入資料
file_path = r'C:\Users\User\Desktop\大學\四上\DM\RLCt13C2_merged_data.csv'

try:
    # 嘗試不同的編碼方式讀取資料，解決中文亂碼問題
    data = pd.read_csv(file_path, encoding='utf-8')  # 常見編碼選項：'utf-8', 'big5', 'gbk'
    print("Data loaded successfully!")
```

```python
# 輸出變數名稱為串列
variable_names = data.columns.tolist()
print("Variable Names:", variable_names)
```

```python
# 預定的變數名稱
new_variable_names = [
    'Township and urban area', 'Transaction subject', 'Land location and building number', 
    'Total land transfer area square meters', 'Urban land use zoning', 
    'Non-urban land use zoning', 'Non-urban land use Edit', 'Transaction date', 
    'Transaction number', 'Transfer level', 'Total number of floors', 
    'Building type', 'Main purpose', 'Main building materials', 
    'Building completion year month', 'Total transferred area of the building in square meters', 
    'Current layout of the building.room', 'Current layout of the building.hall', 
    'Current layout of the building.bathroom', 'Current layout of the building.partitions', 
    'With or without management organization', 'Total price yuan', 
    'Unit price yuan square meter', 'Parking space category', 
    'Total price of parking space yuan', 'Remarks', 'Number', 
    'Main building area', 'Attached Building area', 'Balcony area', 
    'Elevator', 'Transfer number', 'Total parking space transfer area square meters'
]

# 確認新變數名稱數量與資料欄位數量一致
if len(new_variable_names) == len(data.columns):
    data.columns = new_variable_names
    print("Variable names updated successfully.")
else:
    print(f"Error: The number of new variable names ({len(new_variable_names)}) does not match the number of columns ({len(data.columns)}) in the dataset.")

# 顯示更新後的變數名稱
print("Updated Variable Names:", data.columns.tolist())

# 可選：儲存更新後的資料為新的 CSV 檔案
output_file = r'C:\Users\User\Desktop\大學\updated_data.csv'
data.to_csv(output_file, index=False, encoding='utf-8')
print(f"Updated data saved to {output_file}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 檢查資料集中的遺失值
missing_values = data.isnull().sum()

# 分離變數：有遺失值的變數和沒有遺失值的變數
variables_with_missing = missing_values[missing_values > 0].index
variables_without_missing = missing_values[missing_values == 0].index

# 只保留有遺失值的變數
data_with_missing = data[variables_with_missing]

# 對變數（列）按遺失值數量排序
sorted_columns = data_with_missing.isnull().sum().sort_values(ascending=False).index

# 對觀察值（行）按遺失值數量排序
sorted_rows = data_with_missing.isnull().sum(axis=1).sort_values(ascending=False).index

# 重新排序資料
sorted_data = data_with_missing.loc[sorted_rows, sorted_columns]

# 視覺化遺失值模式（熱圖）
plt.figure(figsize=(20, 8))
sns.heatmap(
    sorted_data.isnull(), 
    cbar=False, 
    cmap="viridis", 
    yticklabels=False  # 隱藏 Y 軸刻度
)
plt.title("Missing Value Heatmap (Sorted)")
plt.xlabel("Variables")
plt.xticks(rotation=45)  # 確保 X 軸標籤（變數名稱）不重疊
plt.show()

# 顯示沒有遺失值的變數
if not variables_without_missing.empty:
    print("Variables without missing values:")
    print(variables_without_missing.tolist())
else:
    print("All variables have missing values.")
```
